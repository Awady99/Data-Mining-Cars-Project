```python
features = data.loc[:, 'Buying': 'Safety']
label = data['Overall']
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
res = model.fit(features, label)
```

```python
from sklearn import metrics
```

```python
import numpy as np
```

```python
y_actual=np.asarray(label)
```

```python
y_actual
```

```python
y_predcit = model.predict(features)
```

```python
y_predcit
```

```python
print('Accuracy = ', metrics.accuracy_score(label, y_predcit))
```
